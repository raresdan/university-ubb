```python
x,y = var('x y')
f = (2*x-x^2-x*y,-y+x*y)
jac = jacobian(f,(x,y))
jac
```

```python
A1 = jac.subs(x=0,y=0)
A1
```

```python
A2 = jac.subs(x=2,y=0)
A2
```

```python
A3 = jac.subs(x=1,y=1)
A3
```

```python
A1.eigenvalues()
# saddle - unstable
```

```python
A2.eigenvalues
# saddle - unstable
```

```python
A3.eigenvalues()
# stable - focus
```
