```python
def spinner(probabilityDistribution):
    randAux = np.random.uniform(0, 1)
    index = 0
    while randAux > 0:
        if randAux - probabilityDistribution[index] > 0:
            randAux -= probabilityDistribution[index]
            index += 1
        else:
            return index

def simulate_spinner_trials(probabilityDistribution, n_trials):
    outcomes = []
    for _ in range(n_trials):
        outcome = spinner(probabilityDistribution)
        outcomes.append(outcome)
    return outcomes

def plot_outcomes_histogram(outcomes):
    plt.hist(outcomes, bins=len(set(outcomes)), align='mid', rwidth=0.8, color='blue', edgecolor='black')
    plt.xlabel('Outcome')
    plt.ylabel('Frequency')
    plt.title('Histogram of Spinner Outcomes')
    plt.xticks(range(1, len(set(outcomes)) + 1))
    plt.grid(axis='y', alpha=0.5)
    plt.show()
```

```python
probabilityDistribution1 = [0.2, 0.5, 0.1, 0.2]
n_trials = 100
outcomes = simulate_spinner_trials(probabilityDistribution1, n_trials)
plot_outcomes_histogram(outcomes)
```

```python
pipe_radius = 0.3 
wall = 1
ball_radius = 0.01 
trials = 100
```

```python
def randomPosition():
    x = np.random.uniform(-1/2,1/2)
    y = np.random.uniform(-1/2,1/2)
    return x,y
 
def checkInsidePipe(x,y):
    distanceFromCenter = np.sqrt(x**2+y**2)
    if distanceFromCenter <= pipe_radius:
        return 1
    return 0
    
def runTrials(trials):
    total = 0
    for i in range(0,trials):
        x,y = randomPosition()
        total += checkInsidePipe(x,y)
    return total
 
totalInside = runTrials(trials)
resultedProbability = totalInside / trials
print(resultedProbability)
 
mathematicalProbability = (np.pi * pipe_radius * pipe_radius) / wall**2
print(mathematicalProbability)
```
